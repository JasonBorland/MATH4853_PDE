Use the method of characteristics to obtain the general solution of each of the following partial differential equations.  In each case, sketch some characteristic curves

### Part(i) 
\begin{align}
\frac{\partial u}{\partial x} -\, \frac{\partial u}{\partial y} & = \frac{u}{x + y}
\end{align}


\begin{align}
\frac{dy}{dx}  & = \frac{b}{a} & = \frac{-1}{1} & = -1 & (1)\\
\frac{du}{dx}  & = \frac{c}{a} & = \frac{\frac{u}{x+y}}{1} & =  \frac{u}{x+y} & (2)
\end{align}

Find general solution; Solve ode (1):
\begin{align}
\frac{dy}{dx}  & = -1\\
\therefore dy &= -dx \\
\therefore \int dy &= \int -dx \\
\therefore y &= -x + k \\
\therefore k &= x + y
\end{align}

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-5., 5., 0.2)



fig, ax = plt.subplots(figsize=(10,10))

ax.set_title('Plot of Characteristic Curves in XY-Plane')
ax.set_xlabel('$x$', fontsize=15 )
ax.xaxis.set_label_coords(1, 0.49) 
ax.set_ylabel('$y$', fontsize=15 )
ax.yaxis.set_label_coords(0.49, 1) 

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

#ax.set_ylim(-5,5)
for k in range (-5, 6):
    y = -x + k
    plt.plot(x,y, label='k='+str(k) )
ax.legend(loc='best', fontsize=10)

plt.show()
```

Solve ode (2):
\begin{align}
\frac{du}{dx}  & = \frac{u}{x+y}\\
\therefore \frac{du}{dx}  & = \frac{u}{x+(-x+k)}\\
\therefore \frac{du}{dx}  & = \frac{u}{k}\\
\therefore \frac{du}{u} & = \frac{1}{k} dx\\
\therefore \int \frac{du}{u} & = \frac{1}{k} \int dx\\
\therefore \ln \vert u \vert &= \frac{x}{k} + f(k) \\
\therefore \vert u \vert & = e^{\frac{x}{k} + f(k)} \\
\therefore \vert u \vert & = e^{f(k)}e^{\frac{x}{k}} \\
\therefore u & = \pm e^{f(k)}e^{\frac{x}{k}} = g(k)e^{\frac{x}{k}} \quad \quad g(x) = \pm e^{f(k)}\\
\therefore u &= g(x+y)e^{\frac{x}{x+y}}
\end{align}

### Part(ii) 
\begin{align}
x(x+y)\frac{\partial u}{\partial x} +\, y(x+y)\frac{\partial u}{\partial y} & = (x - y)u
\end{align}



\begin{align}
\frac{dy}{dx}  & = \frac{b}{a} & = \frac{y(x+y)}{x(x+y)}  &= \frac{y}{x} & (1)\\
\frac{du}{dx}  & = \frac{c}{a} & = \frac{u(x-y)}{x(x+y)}  & & (2)
\end{align}

Find general solution; Solve ode (1):
\begin{align}
\frac{dy}{dx}  & = \frac{y}{x}\\
\therefore \frac{dy}{y} &= \frac{dx}{x} \\
\therefore \int \frac{1}{y}dy &= \int \frac{1}{x}dx \\
\therefore \ln \vert y \vert &= \ln \vert x \vert + k \\
\therefore \vert y \vert &= e^{\ln \vert x \vert + k} \\
\therefore \vert y \vert &= e^{k}e^{\ln \vert x \vert} \\
\therefore y  &= \pm e^{k}x \\
\therefore y  &= Lx  & \quad L = \pm e^{k} \\
\therefore L &= \frac{y}{x} 
\end{align}

```python
x = np.arange(-5., 5., 0.2)



fig, ax = plt.subplots(figsize=(10,10))

ax.set_title('Plot of Characteristic Curves in XY-Plane')
ax.set_xlabel('$x$', fontsize=15 )
ax.xaxis.set_label_coords(1, 0.49) 
ax.set_ylabel('$y$', fontsize=15 )
ax.yaxis.set_label_coords(0.49, 1) 

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

#ax.set_ylim(-5,5)
for L in range (-5, 6):
    y = x*L
    plt.plot(x,y, label='L='+str(L) )
ax.legend(loc='best', fontsize=10)

plt.show()
```

Solve ode (2):
\begin{align}
\frac{du}{dx} & = \frac{u(x-y)}{x(x+y)} \\
\therefore \frac{du}{dx} & = \frac{u(x- Lx)}{x(x+Lx)}\\
\therefore \frac{du}{dx} & = u\frac{x(1-L)}{x^2(1+L)}\\
\therefore \frac{du}{dx} & = u\frac{(1-L)}{x(1+L)}\\
\therefore \frac{1}{u}du &= \frac{(1-L)}{x(1+L)} dx \\
\therefore \int \frac{1}{u}du &= \frac{(1-L)}{(1+L)} \int \frac{1}{x}dx \\
\therefore \ln \vert u \vert &= \frac{(1-L)}{(1+L)} \ln \vert x \vert + f(L) \\
\therefore \ln \vert u \vert &= \ln \vert x \vert^{\frac{(1-L)}{(1+L)}}  + f(L) \\
\therefore \vert u \vert &= e^{\ln \vert x \vert^{\frac{(1-L)}{(1+L)}}  + f(L)}\\
\therefore \vert u \vert &= e^{f(L)}e^{\ln \vert x \vert^{\frac{(1-L)}{(1+L)}}}\\
\therefore u &= \pm e^{f(L)}e^{\ln x^{\frac{(1-L)}{(1+L)}}}\\
\therefore u &= f(L)e^{\ln x^{\frac{(1-L)}{(1+L)}}}  & \quad f(L) = \pm e^{f(L)}\\
\therefore u &= f(\frac{y}{x}) x^{\frac{x-y}{x+y}} 
\end{align}
