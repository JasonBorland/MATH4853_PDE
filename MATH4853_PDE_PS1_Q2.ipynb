### Part (i)
\begin{align}
3\frac{\partial u}{\partial x} +\, 4\frac{\partial u}{\partial y} & = \frac{u}{x} \, \quad where \, u(1,y) = y^2 \, for -\infty < y < \infty
\end{align}



\begin{align}
\frac{dy}{dx}  & = \frac{b}{a} & = \frac{4}{3}  &  & (1)\\
\frac{du}{dx}  & = \frac{c}{a} & = \frac{\frac{u}{x}}{3} & =  \frac{u}{3x} & (2)
\end{align}

Find general solution; Solve ode (1):
\begin{align}
\frac{dy}{dx}  & = \frac{4}{3}\\
\therefore dy &= \frac{4}{3} dx \\
\therefore \int dy &= \frac{4}{3} \int dx \\
\therefore y &= \frac{4}{3}x + k \\
\therefore k &= -\frac{4}{3}x + y
\end{align}

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

Solve ode (2):
\begin{align}
\frac{du}{dx}  & = \frac{u}{3x}\\
\therefore \frac{du}{u} & = \frac{1}{3x} dx\\
\therefore \int \frac{du}{u} & = \frac{1}{3} \int \frac{1}{x} dx\\
\therefore \ln \vert u \vert &=  \frac{1}{3} \ln \vert x \vert + f(k)\\
\therefore \ln \vert u \vert &=  \ln \vert x^\frac{1}{3} \vert + f(k)\\
\therefore \vert u \vert & = e^{\ln \vert x \vert^\frac{1}{3} + f(k)} \\
\therefore \vert u \vert & = e^{f(k)} e^{\ln \vert x \vert^\frac{1}{3}} \\
\therefore u  & = \pm e^{f(k)} e^{\ln x^\frac{1}{3}} & = g(k)x^{\frac{1}{3}} & \quad \, where \, g(k) = \pm e^{f(k)}\\
\therefore u &= g(y - \frac{4}{3}x)x^{\frac{1}{3}}
\end{align}



Use the initial conditions $u(1,y) = y^2$ for $-\infty < y < \infty$
\begin{align}

### Part (ii)
\begin{align}
x\frac{\partial u}{\partial x} +\, y\frac{\partial u}{\partial y} & = -x^2u^3 \, \quad where \, u(x,1) = x \, for -\infty < y < \infty
\end{align}



\begin{align}
\frac{dy}{dx}  & = \frac{b}{a} & = \frac{y}{x}  &  & (1)\\
\frac{du}{dx}  & = \frac{c}{a} & = \frac{-x^2u^3}{x} & =  -xu^3 & (2)
\end{align}

Find general solution; Solve ode (1):
\begin{align}
\frac{dy}{dx}  & = \frac{y}{x}\\
\therefore \frac{dy}{y} &= \frac{1}{x} dx \\
\therefore \int \frac{1}{y}dy &=  \int \frac{1}{x} dx \\
\therefore \ln \vert y \vert &= \ln \vert x \vert + k \\
\therefore \vert y \vert &= e^{\ln \vert x \vert + k} \\
\therefore \vert y \vert &= e^k e^{\ln \vert x \vert} \\
\therefore y = \pm e^k x &= L x & \quad where \, L = \pm e^k \\
\therefore L = \frac{y}{x}
\end{align}

Solve ode (2):
\begin{align}
\frac{du}{dx} & = -xu^3 \\
\therefore \frac{du}{u^3} & = -x dx \\
\therefore \int \frac{du}{u^3} & = \int -x dx \\
\therefore -\frac{1}{2}u^{-2} &= \frac{-1}{2}x^2 - \frac{1}{2}f(k) \\
\therefore u^{-2} &= x^2 + f(k) \\
\therefore u^2 &= \frac{1}{x^2 + f(k)} \\
\therefore u = \frac{\pm 1}{\sqrt{x^2 + f(k)}} \\
\therefore u = \frac{\pm 1}{\sqrt{x^2 + f(\frac{y}{x})}}
\end{align}

Initial conditions: $u(x,1) = x \, for -\infty < y < \infty$

\begin{align}
x &= \frac{\pm 1}{\sqrt{x^2 + f(\frac{1}{x})}} \\
\therefore x^2 & = \frac{1}{x^2 + f(\frac{1}{x})} \\
\therefore x^2(x^2 + f(\frac{1}{x})) &= 1 \\
\therefore f(\frac{1}{x}) &= \frac{1 - x^4}{x^2} \\
\therefore f(\frac{y}{x}) &= \frac{1 - \frac{y^4}{x^4}}{\frac{y^2}{x^2}} \\ 
\therefore f(\frac{y}{x}) &= \frac{x^4 - y^4}{x^2y^2} \\
\therefore u &= \frac{\pm 1}{\sqrt{x^2 + \frac{x^4 - y^4}{x^2y^2}}} \\
\therefore u &= \frac{\pm xy}{\sqrt{x^4y^2 + x^4 - y^4}} \\
\end{align}

The domain of influence requires $x^4y^2 + x^4 - y^4 > 0$

Using the parametric form of the Method of characteristics

\begin{align}
\frac{dx}{dt}  & = x  & (1)\\
\frac{dy}{dt}  & = y & (2) \\
\frac{du}{dt}  & = -x^2u^3 & (3) \\
\end{align}

Solving ODE (1)
\begin{align}
\frac{dx}{dt}  & = x \\
\therefore \frac{dx}{x} &= dt \\
\therefore \int \frac{dx}{x} &= \int dt \\
\therefore \ln \vert x \vert &= t + A \\
\therefore \vert x \vert &= e^{t + A} &= e^Ae^t \\
\therefore x &= \pm e^A e^t &= Be^t \quad \quad where \, B=\pm e^A & (3)\\ 
\end{align}

Solving ODE (2)

\begin{align}
\frac{dy}{dt}  & = y \\

\therefore \vert y \vert &= e^{t + C} &= e^Ce^t \\
\therefore y &= \pm e^C e^t &= De^t \quad \quad where \, D=\pm e^C & (4)\\
\end{align}

where $u(x,1) = x \, for -\infty < y < \infty$
