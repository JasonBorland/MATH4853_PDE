### Question 2:
Use the parametric form of the Lagrange-Charpit equations to solve the problem:
\begin{align}
\Big ( \frac{\partial u}{\partial x} \Big ) \Big( \frac{\partial u}{\partial y} \Big ) &= xy
\end{align}
where $u(0,y) = -y^2$ for $ y \in \mathbb{R}$



First lets parameterise the initial conditions:
At $t=0$, $x=0$, $y=s$ and $u=-s^2$.

Our PDE can be written as: $F(x,y,u,p,q) = pq -xy =0$  (1).

The Lagrange-Charpit equations in parametric form are:
\begin{align}
\frac{dx}{dt} &= F_p = q \quad (2)\\
\frac{dy}{dt} &= F_p = p \quad (3)\\
\frac{du}{dt} &= pF_p +qF_q = pq + qp = 2pq  \quad (4)\\
\frac{dp}{dt} &= -F_x -pF_u = -(-y) -p(0) = y \quad (5)\\
\frac{dq}{dt} &= -F_y -qF_u = -(-x) -q(0) = x \quad (6)\\
\end{align}

\begin{align}
dt &= \frac{dx}{q} = \frac{dy}{p} = \frac{du}{2pq} = \frac{dp}{y} = \frac{dq}{x} \\
\end{align}


From (2) and (6), we get:
\begin{align}
\frac{dx}{dt} &= q \quad (2)\\
\therefore \frac{d^2x}{dx^2} &= \frac{dq}{dt} \\
\therefore \frac{d^2x}{dx^2} &= x \quad from \, (6)\\
\therefore x &= Acosht + Bsinht \quad from \,solutions \, to \, standard \, ODEs \\
\therefore 0 &= A(1) + B(0) \quad from \, initial \, conditions \\
\therefore A &= 0, \, B=? \\
\therefore x &= Bsinht  \quad (7) \\
\therefore \frac{dx}{dt} &= Bcosht = q \quad (8)\\
\end{align}

Similarly from (3) and (5):
\begin{align}
\frac{dy}{dt} &= p \quad (3)\\
\therefore \frac{d^2y}{dx^2} &= \frac{dp}{dt} \\
\therefore \frac{d^2y}{dx^2} &= y \quad from  \, (5)\\
\therefore y &= Ccosht + Dsinht \quad from \,solutions \, to \, standard \, ODEs \\
\therefore s &= C(1) +D(0) \quad from \, initial \, conditions \\
\therefore C &= s, D=? \\
\therefore y &= scosht + Dsinht \quad (9)\\
\therefore \frac{dy}{dt} &= ssinht + Dcosht = p \quad (10) \\
\end{align}

Using Equation (1) and (7), (8), (9) and (10) :
\begin{align}
pq &= xy \\
\therefore (ssinht + Dcosht)(Bcosht) &= Bsinht(scosht + Dsinht) \\
\therefore sB(cosht)(sinht) + BDcosh^2t &= sB(cosht)(sinht) + BDsinh^2t \\
\therefore BD &= 0, \, B \ne 0, \, so \, D = 0\\
\therefore y &= scosht \quad (11)\\
\therefore \frac{dy}{dt} &= ssinht = p \quad (12) \\
\end{align}

Now we will substitute (12) and (8) into (4):
\begin{align}
\frac{du}{dt} &= 2pq \\
\therefore \frac{du}{dt} &= 2(ssinh(t))(Bcosh(t)) \quad where \, s, \, and \, B \, are \, constants \\
\therefore u &= 2Bs \int sinh(t)cosh(t) dt = 2Bs \int \frac{sinh(2t)}{2} dt\\
\therefore u &=  \frac{sBcosh(2t)}{2} + E \quad \\
\therefore -s^2 &= \frac{sBcosh2(0)}{2} + E \\
\therefore -s^2 &= \frac{sB}{2} +E \\ 
\therefore B &= -2s, \, E=0  \\
\therefore x &= -2ssinh(t) \\
\therefore y &= scosh(t) \\
\therefore x^2 &= 4s^2sinh^2(t) \\
\therefore y^2 &= s^2cosh^2(t) \\
\therefore u &= -s^2cosh(2t) = -s^2(cosh^2(t) + sinh^2(t)) \\
\therefore u &= -\frac{x^2}{4} - y^2 \\
\end{align}
